```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```python
file = "../Data/README.md"
word = sc.textFile(file).flatMap(lambda x : x.split(" "))
len_word_pair = word.map(lambda x : (len(x),x))
len_word_pair.glom().collect()
```

```python
len_word_pair_group = len_word_pair.groupByKey()
```

```python
len_word_pair_group.collect()
```

## Pairs with a same key tend to in the same partition

```python
len_word_pair = len_word_pair_group.flatMapValues(lambda x : x)
```

```python
len_word_pair.glom().collect()
```

## Other flatmapValues() Examples

```python
x = sc.parallelize([("a", ["x", "y", "z"]), ("b", ["p", "r"])])
```

```python
x.flatMapValues(lambda x : x).glom().collect()
```
