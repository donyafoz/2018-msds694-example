```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```python
file = "../Data/README.md"
```

```python
word = sc.textFile(file).flatMap(lambda x : x.split(" "))#.filter(lambda x : x != '')
```

```python
word.take(5)
```

# Create a key-value pair

```python
len_word_pair = word.map(lambda x : (len(x),x))
```

# Keys

```python
len_word_pair.keys()
```

```python
len_word_pair.keys().collect()
```

# Values

```python
len_word_pair.values()
```

```python
len_word_pair.values().collect()
```

# Sort by key

```python
len_word_pair.sortByKey().take(5)
```

```python
len_word_pair.sortByKey(ascending=False).take(5)
```

# Group by key to create (length of a word, list of words) 

```python
len_word_pair_group = len_word_pair.groupByKey()
```

```python
len_word_pair_group.take(5)
```

```python
len_word_pair_group.map(lambda x : (x[0], list(x[1]))).take(5)
```
