# Create key-value pairs of (word, 1)

```python
from pyspark import SparkContext 
sc = SparkContext.getOrCreate()
```

```python
file_name = "../Data/README.md"
```

```python
word = sc.textFile(file_name).flatMap(lambda x : x.split(" "))
```

```python
word_map = word.map(lambda x : (x,1))
```

```python
word_map.collect()
```

# Extra - Create key-value pairs of (length of the word, 1)

```python
word_map = word.map(lambda x : (len(x), 1))
```

```python
word_map.collect()
```
